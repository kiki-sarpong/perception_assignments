```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.examples.tutorials.mnist import input_data
import tensorflow as tf
import pandas as pd
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
```

```python
fashion_mnist = tf.keras.datasets.fashion_mnist
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
```

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
nrow = test_images.shape[1] * test_images.shape[2]
ncol_test = test_images.shape[0]
ncol_train = train_images.shape[0]
data_test = test_images.reshape(ncol_test,nrow)
data_train = train_images.reshape(ncol_train,nrow)
data_train = data_train/255
data_test = data_test/255

print(data_test.shape)
```

```python
iterations = np.arange(0,10)   #index labels
colors = ['b','r','g','c','#ffa600','y','#003f5c','k','#bc5090','chartreuse']  #different colors

pca = PCA(n_components=2).fit(data_test)   #fit data
pca_result = pca.transform(data_test)
plt.figure('Fashion Plot - PCA',figsize=(10,10))

for color,ind,class_name in zip(colors,iterations,class_names):
    plt.scatter(pca_result[test_labels == ind, 0], pca_result[test_labels == ind,1],color=color,alpha =.8,lw=2,label = class_name)

plt.title("Two Dimensional Representation of Fashion MNIST - PCA")
plt.legend(loc="best")
```

```python
tsne = TSNE(n_components=2, verbose=1, perplexity=5, n_iter=1200)
tsne_results = tsne.fit_transform(data_test)
```

```python
plt.figure('Fashion Plot - TSNE',figsize=(10,10))
for color,ind,class_name in zip(colors,iterations,class_names):
    plt.scatter(tsne_results[test_labels == ind, 0], tsne_results[test_labels == ind,1],color=color,alpha =.8,lw=2,label = class_name)

plt.title("Two Dimensional Representation of Fashion MNIST - TSNE")
plt.legend(loc="best")
```
